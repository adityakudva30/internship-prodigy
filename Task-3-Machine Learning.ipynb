```python
import os
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import zipfile
```

```python
# Specify the path to your zip file
zip_file_path = 'C:\\Users\\Aditya Kudva\\Downloads\\kagglecatsanddogs_3367a.zip'
# Specify the directory where you want to extract its contents
extract_to_path = 'C:\\Users\\Aditya Kudva\\Downloads\\Extracted'

# Create the directory if it doesn't exist
os.makedirs(extract_to_path, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_path)

print("Extraction Complete")
```

```python
# Base directory where the extracted files are located
base_dir = 'C:\\Users\\Aditya Kudva\\Downloads\\Extracted\\PetImages'

categories = ['Cat', 'Dog']

data = []  # This will hold tuples of (image_data, label)

for category in categories:
    path = os.path.join(base_dir, category)
    label = categories.index(category)

    if os.path.exists(path):
        for img in os.listdir(path):
            img_path = os.path.join(path, img)
            try:
                pet_img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)
                if pet_img is not None:
                    pet_img = cv.resize(pet_img, (64, 64))  # Resize to (64, 64)
                    data.append((pet_img.flatten(), label))  # Append a tuple
            except Exception as e:
                print(f"Error loading image {img_path}: {e}")
    else:
        print(f"Directory not found: {path}")

print("Data loading complete.")
```

```python
random.shuffle(data)

features = []
labels = []

for feature, label in data:
    features.append(feature)
    labels.append(label)  
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
x_train, x_test, y_train, y_test
```

```python
model = SVC()
model

```

```python
model.fit(x_train, y_train)
accuracy = model.score(x_test, y_test)
print(f"Test accuracy: {accuracy * 100:.2f}%")
```

```python
mypet = x_test[2].reshape(100,100)
prediction = model.predict(x_test)
accuracy = model.score(x_test,y_test)
print('Accuracy :', accuracy)
print('Prediction is', categories[prediction[2]])
plt.imshow(mypet, cmap ='gray')
plt.show()
```
